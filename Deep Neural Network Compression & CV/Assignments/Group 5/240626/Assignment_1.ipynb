## Part B
### B1. Compute the FFT of the grayscale image ( use NumPy )
Steps:
1. <code>np.fft.fft2</code>
2. <code>np.fft.fftshift</code>
3. Store as <code>fshift</code>

### B2. Compute magnitude, phase and display Magnitude Spectrum

````python

f = np.fft.fft2(gray.astype(np.float32))
fshift = np.fft.fftshift(f)
print("FFT computed:", f.shape)
magnitude = np.abs(fshift)
phase = np.angle(fshift)
print("Magnitude shape:", magnitude.shape)
print("Phase shape:", phase.shape)
magnitude_log = np.log1p(magnitude)
mag_display = (magnitude_log - magnitude_log.min()) / (magnitude_log.max() - magnitude_log.min())
plt.figure(figsize=(6,6))
plt.imshow(mag_display, cmap='gray')
plt.title("Magnitude Spectrum")
plt.axis("off")
plt.show()
````

### C3. Create a square LPF mask
Keep side of square as 50. The square will be at the center of the filter.
Apply, Recontruct and show the output.

````python

def reconstruct_from_fshift_mask(fshift_masked):
    F_unshift = np.fft.ifftshift(fshift_masked)
    recon = np.fft.ifft2(F_unshift)
    recon_real = np.real(recon)
    return np.clip(recon_real, 0, 255).astype(np.uint8)

def circular_mask(shape, radius, lowpass=True):
    H, W = shape
    Y, X = np.ogrid[:H, :W]
    dist2 = (Y - H//2)**2 + (X - W//2)**2
    mask = dist2 <= (radius**2)
    return mask.astype(np.float32) if lowpass else (~mask).astype(np.float32)

radii = [10, 30, 80]
lp_results = []
for r in radii:
    mask = circular_mask(gray.shape, r, lowpass=True)
    fshift_lp = fshift * mask
    recon_lp = reconstruct_from_fshift_mask(fshift_lp)
    lp_results.append((r, mask, recon_lp))
plt.figure(figsize=(12,4))
for i, (r, mask, recon_lp) in enumerate(lp_results):
    plt.subplot(1, len(lp_results), i+1)
    plt.imshow(recon_lp, cmap='gray', vmin=0, vmax=255)
    plt.title(f"LPF radius={r}")
    plt.axis('off')
plt.show()

hp_radius = 30
hp_mask = circular_mask(gray.shape, hp_radius, lowpass=False)
fshift_hp = fshift * hp_mask
recon_hp = reconstruct_from_fshift_mask(fshift_hp)
plt.figure(figsize=(6,5))
plt.imshow(recon_hp, cmap='gray', vmin=0, vmax=255)
plt.title(f"High-pass result (radius={hp_radius})")
plt.axis('off')
plt.show()
side = 50
H, W = gray.shape
mask_sq = np.zeros((H, W), dtype=np.float32)
cy, cx = H//2, W//2
half = side // 2
y0, y1 = max(0, cy - half), min(H, cy + half + (1 if side % 2 else 0))
x0, x1 = max(0, cx - half), min(W, cx + half + (1 if side % 2 else 0))
mask_sq[y0:y1, x0:x1] = 1.0
fshift_sq = fshift * mask_sq
recon_sq = reconstruct_from_fshift_mask(fshift_sq)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(mask_sq, cmap='gray'); plt.title("Square LPF mask (side=50)"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(recon_sq, cmap='gray', vmin=0, vmax=255); plt.title("Square LPF result"); plt.axis('off')
plt.show()
````
